# Logistic_Regression: training

To train your logistic regression function, you will do the following:


You initialize your parameter $\theta$, that you can use in your sigmoid, 
you then compute the gradient that you will use to update $\theta$, and 
then calculate the cost. You keep doing so until good enough.
