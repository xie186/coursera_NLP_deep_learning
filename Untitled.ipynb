# Logistic Regression

## Overview of Logistic Regression

Logistic regression is a statistical model used for binary classification problems. This diagram outlines its core components and flow:




$$
h(x^{(i)}, \theta) = \frac{1}{1 + e^{-\theta^T x^{(i)}}}
$$

```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Generate values for z
z = np.linspace(-6, 6, 100)  # Range of z from -6 to 6
h = sigmoid(z)  # Compute sigmoid for each z

# Plot the sigmoid function
plt.figure(figsize=(8, 6))
plt.plot(z, h, label=r"$h(x^{(i)}, \theta)$")
plt.axhline(0.5, color='gray', linestyle='--', linewidth=0.8)  # Add reference line
plt.title("Sigmoid Function", fontsize=14)
plt.xlabel(r"$\theta^T x^{(i)}$", fontsize=12)
plt.ylabel(r"$h(x^{(i)}, \theta)$", fontsize=12)
plt.grid(alpha=0.3)
plt.legend(fontsize=12)
plt.show()
```
